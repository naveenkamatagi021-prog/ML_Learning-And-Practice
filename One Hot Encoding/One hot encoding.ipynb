```python
#Ordinary Encoding
```

```python
dn=pd.DataFrame({"size":["s","m","l","xl","s","m","l","xl","s","m"]})
```

```python
dn
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
oe.fit(dn[["size"]])
```

```python
oe.transform(dn[["size"]])
```

```python
dn["size_en"]=oe.transform(dn[["size"]])
```

```python
dn
```

```python
#Map Function
```

```python
ord_data1={"s":0,"m":2,"l":5,"xl":3}
```

```python
dn["size"].map(ord_data1)
```

```python
dn["size_en_map"]=dn["size"].map(ord_data1)
```
