```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv(r'C:\Users\Naveen s kamatagi\Downloads\4-hour-course-xl-files\Loan payments data.csv',encoding='latin1')
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=["paid_off_time"])
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

```

```python
ss=StandardScaler()
ss.fit(dataset[["Principal"]])
```

```python
ss.transform(dataset[["Principal"]])
```

```python
dataset["Principal_ss"]=pd.DataFrame(ss.transform(dataset[["Principal"]]),columns=["x"])
```

```python
dataset
```

```python
plt.subplot(1,2,1)
plt.title("Befor")
sns.histplot(dataset["Principal"],kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["Principal_ss"],kde=True)

plt.show()
```

```python
#Normalization (mix_max saclling)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["past_due_days"]])
```

```python
ms.transform(dataset[["past_due_days"]])
```

```python
dataset["past_due_days_ms"]=ms.transform(dataset[["past_due_days"]])
```

```python
dataset
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["past_due_days"],kde=True)


plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["past_due_days_ms"],kde=True)
plt.show()
```

```python
#function Transformer 
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft=FunctionTransformer()
ft.fit(dataset[["age"]])
```

```python
ft.transform(dataset[["age"]])
```

```python
dataset["age_ft"]=ft.transform(dataset[["age"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["age"],kde=True)


plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["age_ft"],kde=True)
plt.show()
```
