```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv(r"C:\Users\Naveen s kamatagi\Downloads\4-hour-course-xl-files\placement.csv", encoding="latin1")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
x.ndim # it will show the dimention of the x
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.predict([[6.89]])
```

```python
lm.score(x_test,y_test)*100
```

```python
y_pred=lm.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset["cgpa"],y_pred,c="red")
plt.legend(["Original data","predict line"])
plt.savefig("predict.png")
plt.show()
```
